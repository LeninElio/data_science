```python
import pandas as pd
```

```python
df = pd.read_csv('./data/djmag.csv')
df
```

```python
columnas = ['anio', 'nombre_dj', 'posicion', 'url']
df.columns = columnas
```

```python
df
```

```python
df[df['nombre_dj'] == 'Tiësto']
```

```python
df['anio'] = df['anio'].astype(str)
df
```

```python
import matplotlib.pyplot as plt
```

```python
nombre_dj = 'Avicii'
df_tiesto = df[df['nombre_dj'] == nombre_dj]

plt.figure(figsize=(12, 6))
plt.plot(df_tiesto['anio'], df_tiesto['posicion'], marker='o', color='red', linestyle='dashed', linewidth=1, markersize=10)

for x, y in zip(df_tiesto['anio'], df_tiesto['posicion']):
    plt.annotate(str(y), (x, y), xytext=(0, 15), textcoords='offset points', ha='center', va='top')

plt.xlabel('Año de la encuesta', fontsize=12)
plt.ylabel('Posición del DJ', fontsize=12)
plt.title(f'Posición de {nombre_dj} por Año', fontsize=16)
plt.show()
```
