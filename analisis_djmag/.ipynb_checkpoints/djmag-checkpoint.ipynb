```python
import pandas as pd
```

```python
df = pd.read_csv('./data/djmag.csv')
df
```

```python
columnas = ['anio', 'nombre_dj', 'posicion', 'url']
df.columns = columnas
```

```python
df
```

```python
df[df['nombre_dj'] == 'Tiësto']
```

```python
import matplotlib.pyplot as plt
```

```python
df_tiesto = df[df['nombre_dj'] == 'Tiësto']
df_tiesto['anio'] = df_tiesto['anio'].astype(str)

plt.figure(figsize=(10, 6))
plt.plot(df_tiesto['anio'], df_tiesto['posicion'])

for x, y in zip(df_tiesto['anio'], df_tiesto['posicion']):
    plt.annotate(str(y), (x, y))

plt.xlabel('Año')
plt.ylabel('Posición')
plt.title(f'Posición de Tiësto por Año')
plt.show()
```
