### Transformaciones de datos: Una transformación es una modificación de la estrutura de nuestra base de datos.

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/infracciones_de_transito.csv')
```

```python
df.head()
```

```python
df['MULTA_INTERNAMIENTO'].unique()
```

```python
df['MULTA_INTERNAMIENTO'].replace('-', 0, inplace=True)
```

```python
df['MULTA_INTERNAMIENTO'] = df['MULTA_INTERNAMIENTO'].astype(float)
```

```python
df.dtypes
```

```python
df['INTERNAMIENTO_INFRACCION'] = df['MONTO_INFRACCION'] + df['MULTA_INTERNAMIENTO']
```

```python
df
```

```python
df['INTERNAMIENTO_INFRACCION'].describe()
```

```python
del(df['INTERNAMIENTO_INFRACCION'])
```

```python
df
```

```python
# Borrar varias columnas
df.drop(['DISTRITO', 'PROVINCIA', 'DEPARTAMENTO'], axis=1, inplace=True)
```

```python
# Se puede eliminar por los indices tambien
df.drop(0)
```
