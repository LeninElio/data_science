### Gestión de la información duplicada o pperdida (missing data).

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/infracciones_de_transito.csv')
```

```python
df.head()
```

```python
# Generamos duplicidad de datos
df_duplicate = df._append(df)
```

```python
df_duplicate
```

```python
df_clean = df_duplicate.drop_duplicates()
```

```python
df_clean
```

```python
len(df_clean) == len(df)
```

```python
df_clean.dropna()
```
