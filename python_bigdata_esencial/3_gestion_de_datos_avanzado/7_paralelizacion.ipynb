## Paralelizaci√≥n de loops en python

```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed
```

```python
df = pd.read_csv("../data/2008.csv", nrows=100000)
df
```

## Implementar una funci√≥n paralelizada

```python
df_sub = df[['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']]
df_sub.head()
```

```python
def retraso_maximo(fila_df):
    if not np.isnan(fila_df).any():
        names = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
        return names[fila_df.index(max(fila_df))]
    else:
        return 'None'
```

```python
results = []
for fila in df_sub.values.tolist():
    results.append(retraso_maximo(fila))
    
results
```

```python
results = Parallel(n_jobs = 1, backend = "multiprocessing")(
    map(delayed(retraso_maximo), df_sub.values.tolist())
)
```
