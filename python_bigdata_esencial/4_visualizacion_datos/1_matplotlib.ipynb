```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/2008.csv', nrows=100000)
df.head()
```

```python
data = np.unique(df['Cancelled'], return_counts=True)
data
```

```python
plt.pie(
    x = data[1], labels = data[0],
    colors=['red', 'green'], shadow=True,
    startangle=90,
    radius=2,
)

plt.show()
```
